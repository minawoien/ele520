```python
x=np.arange(-10,10.5,0.5).reshape(-1,1)
my=1
Sgm=2
p=norm1D(my,Sgm,x)
```

```python
fc=np.array([1,1,1])*0.4
fig, ax=plt.subplots(1,1)
fig.set_facecolor(fc)
ax.plot(x,x*0,'g.')
ax.plot(x,p,'b')
ax.plot(x,p,'r.')
ax.grid(color='gray',linestyle='-.')
ax.tick_params(colors='yellow')
plt.show()
```

My solution:

```python
x1=np.arange(-10,10.5,0.5).reshape(-1,1)
x2=np.arange(-9,10.5,0.5).reshape(-1,1)
x_1, x_2 = np.meshgrid(x1, x2)
my2 = np.array([[1],[1]])
Smg2 = np.array([[5, 3], [3, 5]])
p2 = norm_2D(my2, Smg2, x_1, x_2)
```

```python
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.plot_surface(x_1, x_2, p2)
plt.show()
```
